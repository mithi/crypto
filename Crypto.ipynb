# Formulating a Problem 

So let's make the problem more specific.

#### Given:
1. So there is a service called [HashFlare](https://hashflare.io/) that lets you rent a SHA-256 algorithm miner for a year (assume 52 weeks or 364 days), you can choose how much share of hashrate you get ($1.50 per GH/s)

2. Every day for the whole year you get an automatic payout in BTC but a maintenance fee will be automatically deducted from this given current the exchange rate between BTC and USD. The maintenance fee is (0.0035 USD /(10 GH/s) / 24 hrs) 

#### Important Assumption: 
- Bitcoin mined by the network comes from the block reward only and not from transaction fees. 

#### Find the answer to the following:

-  Will I make money? If so how much? 

As  with all investments, we are never 100% certain, we can only make estimates based on what we currently know and assumption that are hopefully sound. Needless to day, how much money we will gain or lose depends on how much we put in.

# First Things First 

1. How much hashrate will I get if I make an initial investment of `x` USD?
2. How much in USD should I initially invest if I want to have `Y` amount of hash rate?

````python
TenGH_in_TH = 100 # There are one hundred 10 GH/s in one TH/s
    
def usd_payment_given_hashrate_contract(
  hashrate = 3, hashrate_fee = 1.50, maintenance_fee = 0.0035):
  '''
  Computes the a breakdown of what you'd be paying in the USD currency,
  of a 52-week HashFlare contract, given the share of Hashrate 
  (in TH/s) you're purchasing. 
  
  INPUT: 
    hashrate: in TH/s 
    maintenance_fee: in USD per 10 GH/s per 24 hours
    hashrate_fee: in USD per 10 GH/s 
    
  OUTPUT: tuple of floats in currency specified
    (unfront_fee, total_maintenance_fee, total, maintenance_fee_per_day)
  '''

  days_in_contract = 364 # Number of days in a 52-weeks 

  upfront_fee = hashrate_fee * hashrate * TenGH_in_TH
  maintenance_fee_per_day = maintenance_fee * hashrate * TenGH_in_TH 
  total_maintenance_fee = maintenance_fee_per_day * days_in_contract 
  total_investment = upfront_fee + total_maintenance_fee

  return (upfront_fee, total_maintenance_fee,
          total_investment, maintenance_fee_per_day)


def hashrate_given_upfront_payment(payment = 450.0, hashrate_fee = 1.50):
  '''
  INPUT: 
    payment: in USD 
    hashrate_fee: in USD per 10 GH/s 
  
  OUTPUT: 
    Hashrate TH/s for the duration of the contract    
  '''
  return payment / hashrate_fee / TenGH_in_TH


def usd_to_bitcoin_converter(amount = 1.0, exchange_rate = 17790.0):
  ''' 
  Given amount in USD and exchange rate (USD per bitcoin)
  Return BTC equivalent
  
  Note: Current price from coinmarketcap.com
  '''
  return amount / exchange_rate 
````

````python
# How much money am I willing to invest? 
x = 450.0 
usd_per_btc = 17790.0 

print("If I had", x, "USD I am willing to invest, I can buy a contract with a hashrate of: ",
       hashrate_given_upfront_payment(payment = x), "TH/s")

print("Given an exchange rate of", usd_per_btc, "my", x,"USD", "is equivalent to", 
      usd_to_bitcoin_converter(amount = x, exchange_rate = usd_per_btc), "BTC")    
````

````python
# How much hashrate in TH/s do I want to buy? 
h = 3

upfront_payment, fees, total, daily_fee = usd_payment_given_hashrate_contract(
  hashrate = h, hashrate_fee = 1.50, maintenance_fee = 0.0035)

print("If I wanted to buy:")
print(h, "TH/s - SHA-256 cloud mining 1 year contract in HashFlare.")
print()
print("Then:")
print("1. I would pay an upfront fee of", upfront_payment, "USD.")
print("2. Every 24 hours,",  daily_fee, "USD would be deducted from the BTC I have mined")
print("based on the exchange rate at that time.")
print("3. This will amount to a total of", fees, "USD for the span of 52 weeks.")
print("4. In total, the HashFlare cloud mining service will take a total of", total, "USD from me.")
````

 # How much will I mine every 10 minutes? 
 
Theoretically and on average, it is targeted that every 10 minutes, a block is solved by the network of miners. Each time a block is solved, a number of bitcoins is added to the network by being awarded to the miner (or mining pool) who solved it. This number of bitcoins is the block reward which is currently `12.5`. This value will halve every 210,000 blocks, with the next halving still more than 2 years away [See: Bitcoin Block Half](bitcoinblockhalf.com). Estimating your piece of the pie or the fraction you will get is simple; it's your hashrate divided by the total hashrate of all the computers participating in the network. The current network hashrate can be seen [here](https://blockchain.info/charts/hash-rate).

````python
block_reward = 12.5 # Get at: bitcoinblockhalf.com
h = 3 # Your hashrate in TH/s
Hn =  13408415 # Network hashrate in TH/s - Get at: blockchain.info/charts/hash-rate

btc_per_block = block_reward * h / Hn
blocks_per_day = 24 * 6 # Number of 10 minutes per 24 hours

print("Bitcoin you'd mine per day:", btc_per_block * blocks_per_day)
print("Bitcoin you'd mine per block (10 minutes):", btc_per_block)
print("Given the following:")
print("- your hashrate in TH/s : ", h)
print("- current network hashrate in TH/s : ", Hn, " - from: blockchain.info/charts/hash-rate")
print("- block reward in TH/s : ", block_reward, " - from: bitcoinblockhalf.com")
````

# The network hashrate varies over time

The network hashrate ` Hn` and difficulty `D` is related by this equation:

```
 Hn = D * 2**32  / 600 
```

The complicated thing is that the difficulty is adjusted every 2016 blocks (2 weeks) based on the time it takes to find the previous 2016 blocks.

# Also, looking at its history, the difficulty and consequently network hashrate tends to increase exponentially over time

- Let's use this! 

![Trends](./img/hn_d_trend.png)

````python
from math import exp, log

def find_average_difficulty_increase(
    difficulty_start = 0.31, difficulty_end = 1.59, number_of_weeks = 52):
  '''
  input:
     difficulty_start: in T
     difficulty_end: in T
     number_of_weeks: weeks between given start and end difficulty
  
  return: Average increase in decimal
   
  Note: Difficulty is adjusted every 2 weeks
   
  Formula used:
    math.stackexchange.com/questions/1280771/finding-a-base-given-an-exponent

  Must satisfy:
    difficulty_end = difficulty_start * (1 + rate_of_increase) ** (number_of_weeks / 2.0)
  '''

  assert difficulty_end >= difficulty_start
  return (exp(log(difficulty_end / difficulty_start) / (number_of_weeks / 2)) - 1.0)
````

````python
# Get values from blockchain.info/charts/difficulty
difficulty_start = 0.31 # 2016/12/23 in T
difficulty_end = 1.59 # 2017/12/16 in T
number_of_weeks = 52 

rate_of_increase = find_average_difficulty_increase(
    difficulty_start, difficulty_end, number_of_weeks)

print("Initial Difficulty: ", difficulty_start)
print("Current Difficulty: ", difficulty_end)
print("Number of weeks between given difficulties: ", number_of_weeks)
print("Rate of increase is", rate_of_increase, "compounded every two weeks.")

theoretical_difficulty_end = difficulty_start * (1 + rate_of_increase) ** (number_of_weeks / 2.0)

print("------")
print("Given a rate of increase of", rate_of_increase, "every two weeks,")
print("after", number_of_weeks, "weeks, from a starting difficulty of", difficulty_start,",")
print("the final difficulty is", theoretical_difficulty_end)
````

# Okay. We can estimate the increase of difficulty. Can we please estimate how much bitcoin we will mine in a year? 

Yes! I'm glad you asked! 

````python
def compute_btc_mined(h = 3.0, Di = 1.59, rid = 6.49, bR = 12.5):
  '''
  Compute BTC mined until HashFlare contract expires (1 year) 
  
  h - hashrate in Terra Hash per seconds
  Di - initial difficulty in trillions 
  rid - rate of increase in difficulty in percent
  bR - block reward, bitcoin per block for the next 52 weeks/lifetime
  
  Estimated bitcoin mined per day: 
  =  h / Hn * bR * 6 * 24,  where Hn = D * (2 **32) / 600
  
  Simplified:
  =    h * bR * 6 * 24 * 600
     ------------------------
        D * (2**32) 

  From: en.bitcoin.it/wiki/Difficulty#What_network_hash_rate_results_in_a_given_difficulty.3F
  Note: Difficulty increases every 2 weeks or 14 days (2016 blocks)
  '''
  
  k = h * bR * 6 * 24 * 600 / (2**32) * 14 
  R = 1.0 + rid / 100 
  D, running_total  = Di, 0 

  # There are 26 2-week periods
  for _ in range(26): 
    running_total += (k / D) 
    D = R * D 

  return running_total
````

````python
# Hashrate in TH/s
hashrate = hashrate_given_upfront_payment(payment = 450.0)

# Initial difficulty in trillions
initial_difficulty = 1.59   

# Rate of increase in difficulty in percent
rate_of_increase_in_difficulty = 6.49 
# For current difficulty: Blockchain.info/charts/difficulty 
# Use find_average_difficulty_increase() to estimate this increase based on history
# Use a larger number than the computed to be more conservative

block_reward = 12.5 # Bitcoin per block for the lifetime of the contract
# From: Bitcoinblockhalf.com

btc_mined = compute_btc_mined(hashrate, initial_difficulty, rate_of_increase_in_difficulty, block_reward)

print("Expect to mine:", btc_mined, "BTC")
print("Given a mining contract of 52 weeks, with a hashrate of", hashrate, "TH/s")
print(" - a block reward of", block_reward, "BTC for the whole lifetime of the contract")
print(" - a starting difficulty of", initial_difficulty, "trillion,")
print(" - and an assumed rate of increase in difficulty of", rate_of_increase_in_difficulty, "percent every two weeks.") 
````

# Great! Did I actually make a profit? 

Well the thing is, the USD per BTC exchange rate is very volatile and daily maintenance fee is subtracted from your bitcoin mined that day given the exchange rate at that time. So we have to make a few more assumptions about the exchange rate which is difficult to predict. 

#### Assumption A: Constant Exchange Rate Scenario - The USD per BTC doesn't change the whole year (You wish!)
- The simplest case for the lazy! If you think the rate will go up and down but will average out through the year then it's not a very bad assumption.

#### Assumption B: Bitcoin Crash Scenario - Right after you bought the contract, the USD per BTC exchange rate drops to X fraction of what it was and stays flat for the whole year. 
- You might think this would be a ridiculuous assumption, but consider these:
- November 30, 2013: 1 BTC = 1,149.14 USD
- May 5, 2014: 1 BTC = 428.98 USD (dropped to a third in about 6 months!)
- May 7, 2015: 1 BTC = 232 USD (almost halved in about a year!)
- If bitcoin's value halves, then HashFlare will get twice as much of your bitcoins given its maintenance fees are pegged in USD!

````python
definitions_dict = {
    
  "upfront_investment" : "How much are you willing to invest in USD? \n",
    
   "crash_number" : " A number between 0.0 and 1.0. The smaller the number, the worse the crash is." +
      "Consider a Bitcoin crash scenario: Right after you bought the contract," +
      "the USD per BTC exchange rate drops to X fraction of what it was and stays flat for the whole year. \n",
    
    "usd_per_btc" : "Current exchange rate from: coinmarketcap.com \n",
    
    "difficulty_start" : "Get the current difficulty from: blockchain.info/charts/difficulty \n",
    
    "pessimistic_rate and optimistic_rate" : " Rate of difficulty increase in percent. " +
      "The optimistic_increase_rate must be less than pessimistic_rate. The higher the more conservative." +
      "A conservative assumption is probably between 10 and 25. \n",

    "block_reward" : "Get the current block reward from: bitcoinblockhalf.com. Constant for the whole contract (1 year). \n",
    
    "hashflare_hashrate_fee": " 1.50 per 10 GH/s from hashflare.io \n",
    
    "hashflare_maintenance_fee": "0.0035 usd per 10GH/s per 24 hours from hashflare.io \n",
  
    "worst_mined_minus_fees" : "Pessimistically, the amount of bitcoins you'd have after the contract. \n",
    
    "best_btc_return" : "Optimistically your net profit in BTC. \n",

    "worst_fees_btc" : " Pessimistically, the total amount in BTC HashFlare took from you. \n",

    "best_mined_minus_fees" : "Optimistically, the amount of bitcoins you'd have after the contract. \n",
    
    "my_hashrate" : "Your hashrate share in TH/s. \n",

    "upfront_btc" : "How much btc you'd have if you bought btc instead of buying a contract. \n",

    "worst_btc_mined" : "Pessimistically, the BTC you've mined before subtracting the maintenance fee. \n",

    "worst_btc_return" : "Optimistically your net profit in BTC. \n",

    "best_btc_mined" : "Optimistically, the BTC you've mined before subtracting the maintenance fee.  \n",

    "best_fees_btc" : "Optimistically, the total amount in BTC HashFlare took from you. \n"
}
````

````python
def get_param_dict(upfront_investment, crash_number, usd_per_btc,
  difficulty_start, pessimistic_rate, optimistic_rate, block_reward,
  hashflare_hashrate_fee, hashflare_maintenance_fee):
  return {
    'upfront_investment': upfront_investment,
    'crash_number': crash_number, 
    'usd_per_btc': usd_per_btc,
    'difficulty_start': difficulty_start,
    'pessimistic_rate': pessimistic_rate,
    'optimistic_rate': optimistic_rate,
    'block_reward': block_reward,
    'hashflare_hashrate_fee': hashflare_hashrate_fee,
    'hashflare_maintenance_fee': hashflare_maintenance_fee
  }

def get_computed_dict(my_hashrate, worst_btc_mined, best_btc_mined, upfront_btc, worst_fees_btc, best_fees_btc, 
  worst_mined_minus_fees, best_mined_minus_fees, best_btc_return, worst_btc_return):
  return {
    'my_hashrate': my_hashrate,
    'worst_btc_mined': worst_btc_mined,
    'best_btc_mined': best_btc_mined,
    'upfront_btc': upfront_btc,
    'worst_fees_btc': worst_fees_btc,
    'best_fees_btc': best_fees_btc,
    'worst_mined_minus_fees': worst_mined_minus_fees,
    'best_mined_minus_fees': best_mined_minus_fees,
    'best_btc_return': best_btc_return, 
    'worst_btc_return': worst_btc_return
  }
````

````python
def hashflare_profit_in_btc(upfront_investment = 450.0, crash_number = 0.5, usd_per_btc = 17790.0,
               difficulty_start = 1.59, pessimistic_rate = 10.0, 
               optimistic_rate = 6.5, block_reward = 12.5,
               hashflare_hashrate_fee = 1.5, hashflare_maintenance_fee = 0.0035):
  '''
  Returns a tuple containing 
   The profits in BTC from HashFlare cloud mining in:
    1. in the constant exchange rate scenario
    2. in the Bitcoin crash scenario 
  3. A dictionary containing all parameters 
  4. A dictionary contraining all computed values
  
  Given the inputs:
    1. upfront_investment 
    - How much are you willing to invest in USD?
    
    2. crash_number
    - Consider a Bitcoin crash scenario:
      Right after you bought the contract, the USD per BTC
      exchange rate drops to X fraction of what it was and
      stays flat for the whole year.
    - A number between 0.0 and 1.0
    - The smaller the number, the worse the crash is
    
    3. usd_per_btc
    - Current exchange rate from: coinmarketcap.com
    
    4. difficulty_start
    - Get the current difficulty from: blockchain.info/charts/difficulty
    
    5. pessimistic_rate
    6. optimistic_rate
    - optimistic_increase_rate must be less than pessimistic_rate
    - A rate of difficulty increase in percent 
    - The higher the more conservative
    - A conservative assumption is probably between 10 and 25

    7. block_reward
    - Get the current block reward from: bitcoinblockhalf.com
    - Constant for the whole contract (1 year)
    
    8. hashflare_hashrate_fee 
    - 1.50 per 10 GH/s from hashflare.io
    
    9. hashflare_maintenance_fee
    - 0.0035 usd per 10GH/s per 24 hours from hashflare.io
  
  '''
  
  assert optimistic_rate <= pessimistic_rate
  assert optimistic_rate >= 0
  assert pessimistic_rate >= 0
  assert hashflare_hashrate_fee > 0
  assert hashflare_maintenance_fee > 0

  # In TH/s
  my_hashrate = hashrate_given_upfront_payment(payment = upfront_investment)
    
  x, fees, total, daily_fee = usd_payment_given_hashrate_contract(
    hashrate = my_hashrate,
    hashrate_fee = hashflare_hashrate_fee, 
    maintenance_fee = hashflare_maintenance_fee)

  worst_btc_mined = compute_btc_mined(h = my_hashrate, 
    Di = difficulty_start, rid = pessimistic_rate, bR = block_reward)

  best_btc_mined = compute_btc_mined(h = my_hashrate, 
    Di = difficulty_start, rid = optimistic_rate, bR = block_reward)

  upfront_btc = usd_to_bitcoin_converter(amount = upfront_investment, exchange_rate = usd_per_btc)
  worst_fees_btc = usd_to_bitcoin_converter(amount = fees, exchange_rate = usd_per_btc * crash_number)
  best_fees_btc = usd_to_bitcoin_converter(amount = fees, exchange_rate = usd_per_btc)

  worst_mined_minus_fees = worst_btc_mined - worst_fees_btc
  best_mined_minus_fees =  best_btc_mined - best_fees_btc 
  
  if worst_mined_minus_fees < 0: worst_mined_minus_fees = 0.0
  if best_mined_minus_fees < 0: best_mined_minus_fees = 0.0
  
  best_btc_return = best_mined_minus_fees - upfront_btc
  worst_btc_return = worst_mined_minus_fees - upfront_btc

  param_dict = get_param_dict(upfront_investment, crash_number, usd_per_btc,
    difficulty_start, pessimistic_rate, optimistic_rate, block_reward,
    hashflare_hashrate_fee, hashflare_maintenance_fee)
  
  computed_dict = get_computed_dict(my_hashrate, worst_btc_mined, best_btc_mined, upfront_btc, worst_fees_btc, best_fees_btc, 
  worst_mined_minus_fees, best_mined_minus_fees, best_btc_return, worst_btc_return)
    
  return (best_btc_return, worst_btc_return, param_dict, computed_dict)
````

````python
def print_dict(D, title):
  print()
  print("---------------------------------------------------------------")
  print(title)
  print("---------------------------------------------------------------")
 
  for key, value in D.items() :
    print(key, ":", value)

def print_suggestion():
    
  print()
  print("---------------------------------------------------------------")
  print("Important Suggestions!")
  print("---------------------------------------------------------------")

  print("The secret to happiness is low expectations.")
  print()
  print("Suggestion 1:")
  print("- Maybe you've used rates of difficulty increase that are too optimistic?")
  print("optimistic_rate and pessimistic_rate")
  print("- A rate of difficulty increase in percent") 
  print("- The higher the more conservative")
  print("- A conservative assumption is probably between 10 and 25")
  print()
  print("Suggestion 2:")
  print("- Maybe you chose crash_number that is too optimistic?")
  print("- Consider a Bitcoin crash scenario, Right after you bought the contract, the USD per BTC")
  print("  exchange rate drops to X fraction of what it was an stays flat for the whole year.")
  print("- A number between 0.0 and 1.0, the smaller the number, the worse the crash is")
  print("- This is a number between 0.0 and 1.0")
  print("- The smaller the worse the crash is")
  print("- Make it smaller, so you don't get dissapointed.")
  print("- A good assumption is probably between 0.1 and 0.4")
    
def print_evaluation(best, worst):
  
  print()
  print("---------------------------------------------------------------")
  print("The constant rate scenario with optimistic difficulty increase")
  print("---------------------------------------------------------------")

    
  if best > 0:
    print("Best case, you'll successfully net:", best, "BTC!")
  elif best == 0: 
    print("This is stupid. You won't lose or gain anything")
    print("Just put it somewhere else."),
  else: 
    print("You will lose:", abs(best), "BTC. Best case, you have been scammed. I told you so.")
    print("You should have just bought bitcoins with it")
    
  print()
  print("---------------------------------------------------------------")
  print("The bitcoin crash scenario with pessimistic difficulty increase")
  print("---------------------------------------------------------------")

  if worst > 0:
    print("Worst case, You'll successfully net:", worst, "BTC! ")
  elif best == 0: 
    print("You didn't lose or gain anything."),
  else: 
    print("You will lose:", abs(worst), "BTC. Worst case, you have been scammed.")
    print("You should have invested in gold??")
    
  if worst > 0 or best > 0:
    print_suggestion()
    
````

````python
best, worst, param_dict, computed_dict = hashflare_profit_in_btc(
  upfront_investment = 450.0, crash_number = 0.5, usd_per_btc = 17790.0,
  difficulty_start = 1.59, pessimistic_rate = 10.0, 
  optimistic_rate = 6.5, block_reward = 12.5,
  hashflare_hashrate_fee = 1.5, hashflare_maintenance_fee = 0.0035)

print_evaluation(best, worst)

print_dict(param_dict, "PARAMETERS")
print_dict(computed_dict, "COMPUTED VALUES" )
print_dict(definitions_dict, "DEFINITIONS")
````
